```python
# generating random colors (in the rgb format)

rgb = [];
# print(rgb)
def generateColor(x):
    for i in range(0,3):
        x.append(random.random() * 255)
    
    for j in range(0, 3):
        x[j] = x[j] / 255
        
    return x

rekt = generateColor(rgb)

x = [1, 2, 3, 4, 5, 6, 7, 8]
y= [5, 2, 4, 2, 1, 4, 5, 2]
plt.scatter(x, y, label = 'skitscat', color = rekt, marker = '*', s = 100)

plt.xlabel('x')
plt.ylabel('y')
plt.title('InterestinGraph\nCheck it out')
plt.legend()
plt.show()
```
