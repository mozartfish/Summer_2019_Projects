```python
import matplotlib.pyplot as plt
import random
import math
```

```python
days = [1, 2, 3, 4, 5]
sleeping = [7, 8, 6, 11, 7]
eating = [2, 3, 4, 3, 2]
working = [7, 8, 7, 2, 2]
playing = [8, 5, 7, 8, 13]

plt.plot([], [], color = 'c', label = 'Sleeping', linewidth = 5)
plt.plot([], [], color = 'm', label = 'Eating', linewidth = 5)
plt.plot([], [], color = 'k', label = 'Working', linewidth = 5)
plt.plot([], [], color = 'r', label = 'Playing', linewidth = 5)
plt.stackplot(days, sleeping, eating, working, playing, colors = ['c', 'm', 'k', 'r'])


plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('InterestinGraph\nCheck it out')
plt.show()
```
