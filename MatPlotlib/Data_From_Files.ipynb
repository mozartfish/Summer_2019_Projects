```python
import matplotlib.pyplot as plt
import random
import math
import numpy as np
import csv
```

```python
x = []
y = []

with open('C:/Users/Pranav/Desktop/cs_work/Matplotlib/example.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))

plt.plot(x, y, label = 'Loaded from file')        
# plt.xlabel('x')
# plt.ylabel('y')
# plt.legend()
plt.title('Interesting Graph\nCheck it out')
plt.show()
```

```python
# with numpy
x, y = np.loadtxt('example.txt', delimiter = ',', unpack = True)


plt.plot(x, y, label = 'Loaded from file')        
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Interesting Graph\nCheck it out')
plt.show()
```
