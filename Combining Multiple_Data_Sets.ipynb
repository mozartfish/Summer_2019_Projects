```python
# Data Analysis of the 2016 US Presidential Election, Unemployment in Counties, and Minimum wages

# set up for data analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

unemployment_df = pd.read_csv("../Summer_2019_Projects/unemployment_data.csv")
# unemployment_df.head()

minimum_wage_df = pd.read_csv("../Summer_2019_Projects/Minimum_Wage_Data.csv")

act_min_wage = pd.DataFrame()
for name, group in minimum_wage_df.groupby("State"):
    if act_min_wage.empty:
        act_min_wage = group.set_index("Year")[["Low.2018"]].rename(columns={"Low.2018":name})
    else:
        act_min_wage = act_min_wage.join(group.set_index("Year")[["Low.2018"]].rename(columns={"Low.2018":name}))

act_min_wage.head()
min_wage_corr = act_min_wage.replace(0, np.NaN).dropna(axis=1).corr()
min_wage_corr.head()

```

```python
act_min_wage = act_min_wage.replace(0, np.NaN).dropna(axis=1)
act_min_wage.head()
```

```python
# take min wage and add it to county unemployment data set
def get_min_wage(year, state):
    try:
        return act_min_wage.loc[year][state]
    
    except:
        return np.NaN
get_min_wage(2012, "Colorado")
```

```python
%%time
# time will give us the total time to perform some cell's operation.

unemployment_df['min_wage'] = list(map(get_min_wage, unemployment_df['Year'], unemployment_df['State']))
```
