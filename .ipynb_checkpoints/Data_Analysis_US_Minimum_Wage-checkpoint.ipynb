```python
# Data Analysis on the United States minimum wage data set from kaggle

# set up for data analysis
import pandas as pd
```

```python
df = pd.read_csv("../Summer_2019_Projects/Minimum_Wage_Data.csv")
```

```python
df.head()
```

```python
gb = df.groupby("State")
gb.get_group("Alabama").set_index("Year").head()
```

```python
act_min_wage = pd.DataFrame()
for name, group in df.groupby("State"):
    if act_min_wage.empty:
        act_min_wage = group.set_index("Year")[["Low.2018"]].rename(columns = {"Low.2018":name})
    
    else:
        act_min_wage = act_min_wage.join(group.set_index("Year")[["Low.2018"]].rename(columns = {"Low.2018": name}))
        

act_min_wage.head()
```

```python
issue_df = df[df['Low.2018'] == 0]
issue_df.head()
```

```python
import numpy as np
act_min_wage.replace(0, np.NaN)
```

```python
act_min_wage.replace(0, np.NaN).dropna(axis = 1).corr().head()
```

```python
min_wage_corr = act_min_wage.replace(0, np.NaN).dropna(axis = 1).corr().head()
for problem in issue_df['State'].unique():
    if problem in min_wage_corr.columns:
        print("missing data")
```
